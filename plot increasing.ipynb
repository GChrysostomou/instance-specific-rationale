```python
import json

fname = "faithfulness_metrics/sst/topk-test-increasing-feature-scoring.json"

with open(fname, "r") as file : data = json.load(file) 
```

```python
data["fixed"]
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.DataFrame(data["fixed"])
```

```python
plt.style.use('tableau-colorblind10')
# plt.rcParams['font.family'] = 'sans-serif'
# plt.rcParams['font.sans-serif'] = 'Helvetica'
plt.rcParams['axes.edgecolor']='#333F4B'
plt.rcParams['axes.linewidth']=0.2
plt.rcParams['xtick.color']='#333F4B'
plt.rcParams['ytick.color']='#333F4B'
plt.rcParams["font.variant"] = "small-caps"

fig, ax = plt.subplots(figsize = (16,10))

df_temp = df.T[["f1 score - model labels"]]
df_temp.plot(ax = ax, marker = "o")
plt.xticks(
    range(len(df_temp.index)), 
    ["\n".join(x.split("-")) for x in df_temp.index], 
    rotation='vertical',
    fontsize = 23
)
plt.yticks(fontsize = 23)
ax.set_ylabel("F1 macro", fontsize = 25)
ax.set_xlabel("Feat-Attr Combo", fontsize = 25)

plt.grid()
plt.tight_layout()


# ax.set_xticklabels(["\n".join(x.split("-")) for x in df_temp.index],  fontsize = 23)
```

```python
plt.style.use('tableau-colorblind10')
# plt.rcParams['font.family'] = 'sans-serif'
# plt.rcParams['font.sans-serif'] = 'Helvetica'
plt.rcParams['axes.edgecolor']='#333F4B'
plt.rcParams['axes.linewidth']=0.2
plt.rcParams['xtick.color']='#333F4B'
plt.rcParams['ytick.color']='#333F4B'
plt.rcParams["font.variant"] = "small-caps"

fig, ax = plt.subplots(figsize = (16,10))

df_temp = df.T[["sufficiency"]]

plt.xticks(
    range(len(df_temp.index)), 
    ["\n".join(x.split("-")) for x in df_temp.index], 
    rotation='vertical',
    fontsize = 23
)
plt.yticks(fontsize = 23)
ax.set_ylabel("F1 macro", fontsize = 25)
ax.set_xlabel("Feat-Attr Combo", fontsize = 25)

plt.grid()
plt.tight_layout()

means = []
stds = []

for _j_, (indx, row) in enumerate(df_temp.iterrows()):
    
    means.append(row["sufficiency"]["mean"])
    stds.append(row["sufficiency"]["std"])

import numpy as np
means = np.asarray(means)
stds = np.asarray(stds)
    
plt.plot(range(len(means)), means, "bo--", markersize = 15)
plt.fill_between(range(len(means)), means - stds*0.1, means + stds*0.1, alpha = 0.2)
```
